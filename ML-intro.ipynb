# Practical Analysis

## Importing the required libraries

```python
import numpy as np ## linear algebra
import pandas as pd ## data processing, data transformation, data cleaning
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt ## used to plot the data
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_percentage_error, mean_absolute_error
```

## Gathering all the required data and coverting it into dataframe

```python
df = pd.read_csv("salary_dataset.csv")
```

## Performing descriptive analysis

```python
df.info()
```

## Performing Statistical Analysis

```python
df.describe()
```

## Checking null values

```python
df.isnull().sum()
```

```python
df.head() #first 5 values of the dataset
```

Defining the Features(X) and Target(Y) for my data

```python
X = df.drop(['Salary', 'Unnamed: 0'], axis=1) #Features
Y = df["Salary"] #Target Variable
```

Splitting the dataset into training and testing part


```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state=42) #test set size of 20%
```

Checking the Split up

```python
X_test.shape
```

### Model Selection

```python
model = LinearRegression()
```

Train Linear Regression model

```python
model = model.fit(X_train, Y_train)
```

Predicting values from the model

```python
predictedSalary = model.predict(X_test)
```

```python
predictedSalary
```

Compared the actual values

```python
errors = mean_squared_error(Y_test, predictedSalary)
print("Mean  Squared Error: ", errors)

r2 = r2_score(Y_test, predictedSalary)
print("R-Squared: ",r2)

mae  = mean_absolute_error(Y_test, predictedSalary)
print("Mean Absolute Error: ",mae)

mape = mean_absolute_percentage_error(Y_test, predictedSalary)
print("Mean Absolute Percentage Error: ", mape)
```

```python
plt.scatter(X,Y, color = 'blue')
plt.plot(X, model.predict(X), color='red')
```
